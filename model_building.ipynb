```python
import pandas as pd
df = pd.read_csv("input.csv")
df.head()
```

```python
# Converting text column into the numeric column using one hot encoding (pandas dummies method)
dummies = pd.get_dummies(df.location, drop_first=True)
dummies.head()
```

```python
# Appending the dummy columns to the main data frame
df = pd.concat([df, dummies], axis="columns")
df.head()
```

```python
df = df.drop(["location"], axis="columns")
df.head()
```

```python
# Creating dependent and independent data frames to feed to the model
X = df.drop(["price"], axis="columns")
X.head()
```

```python
y = df.price
y.head()
```

```python
# Using K fold cross-validation, to train the Linear Regression Model
from sklearn.model_selection import ShuffleSplit
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LinearRegression
cv = ShuffleSplit(n_splits=5, test_size=0.2, random_state=0)
scores = cross_val_score(LinearRegression(), X, y, cv=cv)
scores
```

```python
# Training other models, to see which one performs the best using GridSearchCV
from sklearn.model_selection import GridSearchCV

from sklearn.linear_model import Lasso
from sklearn.tree import DecisionTreeRegressor

def find_best_model_using_gridsearchcv(X,y):
    algorithms = {
        'linear_regression' : {
            'model': LinearRegression(),
            'params': {
                'fit_intercept': [True, False]
            }
        },
        'lasso': {
            'model': Lasso(),
            'params': {
                'alpha': [1,2],
                'selection': ['random', 'cyclic']
            }
        },
        'decision_tree': {
            'model': DecisionTreeRegressor(),
            'params': {
                'criterion' : ['squared_error','friedman_mse'],
                'splitter': ['best','random']
            }
        }
    }
    scores = []
    cv = ShuffleSplit(n_splits=5, test_size=0.2, random_state=0)
    for algo_name, config in algorithms.items():
        gs =  GridSearchCV(config['model'], config['params'], cv=cv, return_train_score=False)
        gs.fit(X,y)
        scores.append({
            'model': algo_name,
            'best_score': gs.best_score_,
            'best_params': gs.best_params_
        })

    return pd.DataFrame(scores,columns=['model','best_score','best_params'])
find_best_model_using_gridsearchcv(X,y)
```

```python
# We see that Linear Regression performs the best
```
