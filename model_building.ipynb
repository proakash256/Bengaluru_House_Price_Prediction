```python
import pandas as pd
df = pd.read_csv("input.csv")
df.head()
```

```python
# Converting text column into the numeric column using one hot encoding (pandas dummies method)
dummies = pd.get_dummies(df.location, drop_first=True)
dummies.head()
```

```python
# Appending the dummy columns to the main data frame
df = pd.concat([df, dummies], axis="columns")
df.head()
```
